## Wrangled Objects to float

```python
df_num = df_drop.copy()

def convert_to_num(column):
    j = []
    for i in column:
        i = str(i)
        a = i.replace(',','')
        a = a.replace('%','')
        a = a.replace('€', '')
        j.append(float(a))
        
    return j
```

```python
col_to_conv = df_num.columns[3:]

for i in col_to_conv:
    df_num[i] = convert_to_num(df_num[i])


df_num.info()
```

## Vizualize whatever you like via seaborn

```python
import seaborn as sns
sns.set()
```

```python
def sort_and_plot(column):
    df_country = df_num.groupby('Shop Country').sum()
    sorted_orders = df_country[column]
    sorted_orders = sorted_orders.sort_values(ascending=False)
    sns.barplot(df_country.index,column, data=df_country, order=sorted_orders.index).set_title(f"Highest : {sorted_orders[0]} by {sorted_orders.index[0]} and Lowest: {sorted_orders.index[-1]} by {sorted_orders.index[-1]}")
```

```python
sort_and_plot('# Orders')      
#copy the line and find profitable countries with 
#order level and absolute terms
```

```python
sort_and_plot('Cost of good Sold')   
```
